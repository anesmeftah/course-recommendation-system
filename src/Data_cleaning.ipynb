```python
# Check for duplicated Course Names
duplicated_courses = courses[courses.duplicated('Course Name', keep=False)]
duplicated_courses[['Course Name']]
```

```python
duplicated_courses.head(5) #Check if there any difference
```

```python
#delete duplicated courses
courses = courses.drop_duplicates('Course Name', keep='first')
```

```python
courses.head()
```

```python
courses.info()
```

```python
courses.insert(0 , 'ID' , range(0,3416))
courses.head()
```

```python
mapping = {'Beginner' : 0 ,
           'Intermediate' : 1,
           'Advanced' : 2}

courses["Difficulty Level"] = courses["Difficulty Level"].map(mapping)

courses.head()
```

```python
courses['Course Name'] = courses['Course Name'].str.lower()
courses['University'] = courses['University'].str.lower()
courses['Course Description'] = courses['Course Description'].str.lower()
courses['Skills'] = courses['Skills'].str.lower()
```

```python
courses.head()
```

```python
courses.to_csv("../Data/cleaned_courses.csv")
```
